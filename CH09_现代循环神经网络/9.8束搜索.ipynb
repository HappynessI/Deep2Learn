# 9.8 束搜索

本节介绍三个搜索策略：
- 贪心搜索(greedy search)
- 穷举搜索(exhaustive search)
- 束搜索(beam search)

## 贪心搜索

- 存在的问题：无法保证得到最优序列
- 原因：在当前时间步做出的决策会影响到后续时间步的条件概率
- 场景：追求计算成本最小

## 穷举搜索

- 存在的问题：计算量可能过大
- 场景：追求精度

## 束搜索

- 说明：介于上述两者之间
- 使用方法：设置超参数束宽k，在时间步1，选择条件概率最大的k个词元。这k个词元将分别是k个候选输出序列的第一个词元。在随后的每个时间步，基于上一个时间步的k个候选输出序列，我们将继续从k|Y|个可能的选择中挑选出条件概率最大的k个候选输出序列
