## 9.4.2 双向模型

双向循环神经网络(bidirectional RNN)添加了反向传递信息的隐藏层，以便更灵活地处理此类信息

设计思路参考了隐马尔可夫模型中的动态规划的前向传递和后向传递

**模型的计算成本及其应用**

这个模型对于对下一个词元进行预测的情况不是所需要的，因为在预测下一个词元时，我们无法知道下一个词元的下文是什么，所以不会得到很高的精确度

这个模型还有一个问题就是：计算速度非常慢。主要原因在于网络的前向传播需要在双向层中进行前向递归和后向递归，并且网络的反向传播还依赖前向传播的结果，因此，梯度求解会经历一个非常长的链。

双向层的使用很少，仅应用于部分场景。例如：填充缺失的单词、词元注释（例如用于命名实体类别）以及作为序列处理流水线中的一个步骤对序列进行编码（例如用于机器翻译）

**总结**
- 在双向循环神经网络中，每个时间步的隐状态由当前时间步的前后数据同时决定
- 双向循环神经网络与概率图模型中的“前向-后向”算法具有相似性
- 双向循环神经网络主要用于序列编码和给定双向上下文的观测估计
- 由于梯度链更长，因此双向循环神经网络的训练成本非常高
